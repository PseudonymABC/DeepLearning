```python
vdata=[]
hdata=[]
vpool=[]
hpool=[]

def convo_img(numimg, filter):             # numimgとfilter には何を入れるんだ？ → n0~n3 画像(12,12)のarray, 　　filterは(3,3)のarrayか？
  nx, ny = len(numimg), len(numimg[0])   # len(numimg)がrowsの数 (=nx)、  　　　　len(numimg[0])がcolumnsの数(=ny)。
  img = np.zeros((nx,ny))                  # 空の12*12のarrayを作る
  for i in range(nx - 3+1):                 #  (画像の高さ) - (フィルタの高さ)+1  .    How did they decide the range? (3*3の形をしたフィルタだからでは？)
    for j in range(ny -3+1):                #  (画像の幅)  -  (フィルタの幅) +1  .    How did they decide the range? (3*3の形をしたフィルタだからでは？)
      img[i][j] =np.sum(numimg[ i : i+3, j : j+3] *filter) #ここは何をしてるんだ？
  return img

def pool_img(numimg, num): #numには何が入るの？
  img=[]
  numimg =np.array(numimg)
  nx, ny = len(numimg), len(numimg[0])  #numimg[0]
  for i in range(0, nx, num):
    row =[]
    for j in range(0, ny, num):
      row.append(np.max(numimg[ i : i+num, j : j+num ]))
    img.append(row)
  return img
```

```python
def pool_img(numimg, num): #numには何が入るの？ー＞　　「pooling層=kernel」の大きさ
  img=[]
  numimg =np.array(numimg)
  nx, ny = len(numimg), len(numimg[0])  #numimg[0]
  for i in range(0, nx, num):
    row =[]
    for j in range(0, ny, num):
      row.append(np.max(numimg[ i : i+num, j : j+num ]))
    img.append(row)
  return img

plt.imshow(np.array(pool_img(n2,3)), "Greys")
# np.array(pool_img(n0,3))
```

```python
def cnn_test(data, num, size):
  # data : ndata = [n0,n1,n2,n3]
  # num : 3  ndata内の画像の数  -> 4でもいいのでは？
  # size : 3  poolingの時のサイズ
  vdata=[]
  hdata=[]
  vpool=[]
  hpool=[]
  for indx  in range(num):
    vdata.append(convo_img(data[indx], fV))
    hdata.append(convo_img(data[indx], fH))
    vpool.append(pool_img(vdata[indx], size))
    hpool.append(pool_img(hdata[indx], size))
  # print(hpool)
  plt.figure(figsize = (12,8))
  for idx in range(num):
    for i in range(len(Filter)):
      plt.subplot(num, 5 , idx*5+i+1)  # 5?
      if i == 0:
        plt.imshow(data[idx], cmap="Greys")
        plt.xlabel(f"Original")
      if i == 1:
        plt.imshow(vdata[idx], cmap="Blues")
        plt.xlabel(f"V")
      if i == 2:
        plt.imshow(vpool[idx], cmap="Blues")
        plt.xlabel(f"V pool")
      if i == 3:
        plt.imshow(hdata[idx], cmap="Blues")
        plt.xlabel(f" H")
      if i == 4:
        plt.imshow(hpool[idx], cmap="Blues")
        plt.xlabel(f"H pool")
      plt.xticks([])
      plt.yticks([])
  plt.show()

cnn_test(ndata, 4,2)
```
