```python
import numpy as np
import matplotlib.pyplot as plt

# make step function
def step_function(x):
  return np.where(x>0, 1,0)

x1 = np.linspace(-5,5,500)
y1 = step_function(x1)

plt.plot(x1,y1)
plt.legend(["step"], loc="best")
plt.yticks(np.arange(0, 1.2, step= 0.5))
plt.grid()
plt.show()
```

```python
# sigmoid function
def sigmoid_func(x):
  return 1/(1+np.exp(-x))

x2 = np.linspace(-5,5)
y2 = sigmoid_func(x2)

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.legend(["step", "sigmoid"], loc="best")
plt.yticks(np.arange(0, 1.2, step= 0.5))
plt.grid()
plt.show()
```

```python
# tanh function
def tanh_fuction(x):
  return np.tanh(x)

x3 = np.linspace(-5,5)
y3 = tanh_fuction(x2)

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.legend(["step", "sigmoid" , "tanh"], loc="best")
plt.yticks(np.arange(0, 1.2, step= 0.5))
plt.grid()
plt.show()
```

```python
# ReLU function
def relu_fuction(x):
  return np.where(x>0, x,0)    #y=x (when x>0)になっている。

x4 = np.linspace(-5,5)
y4 = relu_fuction(x2)

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(x4,y4)
plt.legend(["step", "sigmoid" , "tanh" , "relu"], loc="best")
plt.yticks(np.arange(0, 1.2, step= 0.5))
plt.grid()
plt.show()
```
