```python


X1 = [0,1,0,1]
X2 = [0,0,1,1]

def test(x1,x2):
  if x1 ==1 and x2 ==0:
    return 1
  else:
    return 0

def disp_results(func):
  for i in range(4):
    Y = func(X1[i], X2[i])
    print(f"{X1[i]}, {X2[i]}  = {Y}")

disp_results(test)
```

```python
# 上と同じ処理を、パーセプトロンで作ってみる

def and_test(x1, x2):
  w1, w2 ,theta = 0.5, 0.5, 0.8    # w1, w2, thetaは実際はどう決まるんや？
  ans = w1 * x1 + w2* x2        #
  if ans >theta:
    return 1
  else:
    return 0
disp_results(and_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def fillscolors(data):
    return "#ffc2c2" if data > 0 else "#c6dcec"
def dotscolors(data):
    return "#ff0e0e" if data > 0 else "#1f77b4"


def plot_perceptron(func, X1, X2):
    plt.figure(figsize=(6, 6))
    XX, YY = np.meshgrid(
        np.linspace(-0.25, 1.25, 200),
        np.linspace(-0.25, 1.25, 200))
    XX = np.array(XX).flatten()
    YY = np.array(YY).flatten()
    fills = []
    colors = []
    for i in range(len(XX)):
        fills.append(func(XX[i], YY[i]))
        colors.append(fillscolors(fills[i]))
    plt.scatter(XX, YY, c=colors)

    dots = []
    colors = []
    for i in range(len(X1)):
        dots.append(func(X1[i], X2[i]))
        colors.append(dotscolors(dots[i]))
    plt.scatter(X1, X2, c=colors)
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()
```

```python
plot_perceptron(and_test, X1,X2)
```

```python
# 上と同じ処理を、パーセプトロンで作ってみる

def or_test(x1, x2):
  w1, w2 ,theta = 0.5, 0.5, 0.2    # w1, w2, thetaは実際はどう決まるんや？ ＊＊＊＊thetaのみ　0.8-> 0.2 ！！！！
  ans = w1 * x1 + w2* x2        #
  if ans >theta:
    return 1
  else:
    return 0
disp_results(or_test)
```

```python
plot_perceptron(or_test, X1,X2)
```

```python
# nandを作成
def nand_test(x1, x2):
  w1, w2 ,theta = -0.5, -0.5,- 0.6    # w1, w2, thetaは実際はどう決まるんや？
  ans = w1 * x1 + w2* x2        #
  if ans >theta:
    return 1
  else:
    return 0
disp_results(nand_test)
plot_perceptron(nand_test, X1,X2)
```

```python
# asonn de miru
weight1  =   0.8
weight2  =  -0.8
THETA   =   0
def asobuyou_test(x1, x2):
  w1, w2 ,theta = weight1, weight2, THETA
  ans = w1 * x1 + w2* x2        #
  if ans >theta:
    return 1
  else:
    return 0
disp_results(asobuyou_test)
plot_perceptron(asobuyou_test, X1,X2)
```
